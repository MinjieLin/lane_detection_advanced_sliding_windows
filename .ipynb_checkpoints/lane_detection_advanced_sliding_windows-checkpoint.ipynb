```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import numpy as np

import math
import cv2
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def frame_read(path):
    return mpimg.imread(path)

def frame_show(frame):
    print(type(frame), frame.shape)
    plt.imshow(frame)

def frame_save(frame, path):
    plt.imsave(path, frame)
```

```python
path = 'test_images/straight_lines1.jpg'

frame_origin = frame_read(path)
frame_show(frame_origin)
```

```python
def frame2gray(frame):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    # Or use BGR2GRAY if you read an image with cv2.imread()    
    return cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)

def frame2HLS(frame):
    """Applies the HLS transform"""
    return cv2.cvtColor(frame, cv2.COLOR_RGB2HLS)
```

```python
plt.imshow(frame2gray(frame_origin), cmap='gray')
```

```python
plt.imshow(frame2HLS(frame_origin))
plt.imshow(frame2HLS(frame_origin)[:,:,0])
plt.imshow(frame2HLS(frame_origin)[:,:,1])
plt.imshow(frame2HLS(frame_origin)[:,:,2])
```

```python
plt.imshow(frame2HLS(frame_origin))
plt.figure()

plt.imshow(frame2HLS(frame_origin)[:,:,0])
plt.figure()

plt.imshow(frame2HLS(frame_origin)[:,:,1])
plt.figure()

plt.imshow(frame2HLS(frame_origin)[:,:,2])
plt.figure()
```

```python
frame_HLS_1 = frame2HLS(frame_origin)
plt.imshow(frame_HLS_1)
```

```python
threshold_HLS = (100,75,100)

frame_H = frame_HLS_1[:,:,0]
frame_L = frame_HLS_1[:,:,1]
frame_S = frame_HLS_1[:,:,2]
          
frame_binary = np.zeros_like(frame_S, dtype=bool)
frame_binary[(frame_S > threshold_HLS[2]) & (frame_H < threshold_HLS[0])  & (frame_L > threshold_HLS[1])] = True
```

```python
plt.figure()
plt.imshow(frame_binary)
```
